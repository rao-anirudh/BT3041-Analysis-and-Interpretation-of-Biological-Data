```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("khan_train.csv").transpose().drop(labels="Unnamed: 0",axis=0)
```

```python
df.shape
```

```python
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
Z = linkage(
    y = df,       
    method='complete', 
    metric='euclidean',
    optimal_ordering = True,
)

plt.figure(figsize=(10, 5),dpi=150)
dendrogram(Z,labels=df.index)
plt.title("Euclidean distance with complete linkage");
```

```python
Z = linkage(
    y = df,       
    method='average', 
    metric='euclidean',
    optimal_ordering = True
)

plt.figure(figsize=(10, 5),dpi=150)
dendrogram(Z,labels=df.index)
plt.title("Euclidean distance with average linkage");
```

```python
Z = linkage(
    y = df,       
    method='single', 
    metric='euclidean',
    optimal_ordering = True
)

plt.figure(figsize=(10, 5),dpi=150)
dendrogram(Z,labels=df.index)
plt.title("Euclidean distance with simple linkage");
```

```python
Z = linkage(
    y = df,       
    method='centroid', 
    metric='euclidean',
    optimal_ordering = True
)

plt.figure(figsize=(10, 5),dpi=150)
dendrogram(Z,labels=df.index)
plt.title("Euclidean distance with centroid linkage");
```

```python
Z = linkage(
    y = df,       
    method='complete', 
    metric='cityblock',
    optimal_ordering = True
)

plt.figure(figsize=(10, 5),dpi=150)
dendrogram(Z,labels=df.index)
plt.title("Manhattan distance with complete linkage");
```

```python
Z = linkage(
    y = df,       
    method='average', 
    metric='cityblock',
    optimal_ordering = True
)

plt.figure(figsize=(10, 5),dpi=150)
dendrogram(Z,labels=df.index)
plt.title("Manhattan distance with average linkage");
```

```python
Z = linkage(
    y = df,       
    method='single', 
    metric='cityblock',
    optimal_ordering = True
)

plt.figure(figsize=(10, 5),dpi=150)
dendrogram(Z,labels=df.index)
plt.title("Manhattan distance with simple linkage");
```

```python
Z = linkage(
    y = df,       
    method='centroid', 
    metric='cityblock',
    optimal_ordering = True
)

plt.figure(figsize=(10, 5),dpi=100)
dendrogram(Z,labels=df.index);
```
