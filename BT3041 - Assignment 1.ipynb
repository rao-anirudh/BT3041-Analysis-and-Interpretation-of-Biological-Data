# 5

Calculate summary statistics (mean, median, mode and standard deviation) of chlorophyll a and chlorophyll b measurements from Deciduous forests separately, Evergreen forests separately and both the forests together.

```python
statistics = {"mean":np.mean,"median":np.median,"mode":st.mode,"standard deviation":stats.stdev}
```

<b>Deciduous forests</b>

```python
for statistic in statistics.keys():
    if statistic == "mode":
        chl_a_max_freq = list(st.mode(deciduous_a))[1][0]
        chl_b_max_freq = list(st.mode(deciduous_b))[1][0]
        if chl_a_max_freq == 1:
            print("The mode of chlorophyll a in deciduous forests does not exist as all values occur with the same frequency.")
        if chl_b_max_freq == 1:
            print("The mode of chlorophyll b in deciduous forests does not exist as all values occur with the same frequency.")
            print("\n")
        if chl_a_max_freq != 1 and chl_b_max_freq != 1:
            print(f"The {statistic} of chlorophyll a in deciduous forests is {statistics[statistic](deciduous_a)}.")
            print(f"The {statistic} of chlorophyll b in deciduous forests is {statistics[statistic](deciduous_b)}.")
            print("\n")
    else:    
        print(f"The {statistic} of chlorophyll a in deciduous forests is {statistics[statistic](deciduous_a):.5f}.")
        print(f"The {statistic} of chlorophyll b in deciduous forests is {statistics[statistic](deciduous_b):.5f}.")
        print("\n")
```

<b>Evergreen forests</b>

```python
for statistic in statistics.keys():
    if statistic == "mode":
        chl_a_max_freq = list(st.mode(evergreen_a))[1][0]
        chl_b_max_freq = list(st.mode(evergreen_b))[1][0]
        if chl_a_max_freq == 1:
            print("The mode of chlorophyll a in evergreen forests does not exist as all values occur with the same frequency.")
        if chl_b_max_freq == 1:
            print("The mode of chlorophyll b in evergreen forests does not exist as all values occur with the same frequency.")
            print("\n")
        if chl_a_max_freq != 1 and chl_b_max_freq != 1:
            print(f"The {statistic} of chlorophyll a in evergreen forests is {statistics[statistic](evergreen_a)}.")
            print(f"The {statistic} of chlorophyll b in evergreen forests is {statistics[statistic](evergreen_b)}.")
            print("\n")
    else:    
        print(f"The {statistic} of chlorophyll a in evergreen forests is {statistics[statistic](evergreen_a):.5f}.")
        print(f"The {statistic} of chlorophyll b in evergreen forests is {statistics[statistic](evergreen_b):.5f}.")
        print("\n")
```

# 7

Are the variances between chlorophyll a and chlorophyll b measurements differ significantly? Perform appropriate statistical tests to support your claim. Compare variances of chlorophyll content from Deciduous forests separately, Evergreen forests separately and both the forests together.

<b>Deciduous forests</b>

```python
f_statistic = stats.variance(deciduous_a)/stats.variance(deciduous_b)
df1 = len(deciduous_a) - 1
df2 = len(deciduous_b) - 1
f_critical = st.f.ppf(1-0.05,df1,df2)
p_value = 1 - st.f.cdf(f_statistic,df1,df2)
if f_statistic < f_critical:
    print("The variances of the two groups are NOT significantly different.")
else:
    print("The variances of the two groups are significantly different.")
print(f"F-statistic: {f_statistic:.5f}")
print(f"Critical F-value: {f_critical:.5f}")
print(f"p-value: {p_value}")
```

<b>Evergreen forests</b>

```python
f_statistic = stats.variance(evergreen_a)/stats.variance(evergreen_b)
df1 = len(evergreen_a) - 1
df2 = len(evergreen_b) - 1
f_critical = st.f.ppf(1-0.05,df1,df2)
p_value = 1 - st.f.cdf(f_statistic,df1,df2)
if f_statistic < f_critical:
    print("The variances of the two groups are NOT significantly different.")
else:
    print("The variances of the two groups are significantly different.")
print(f"F-statistic: {f_statistic:.5f}")
print(f"Critical F-value: {f_critical:.5f}")
print(f"p-value: {p_value}")
```

<b>Both forests</b>

```python
f_statistic = stats.variance(both_a)/stats.variance(both_b)
df1 = len(both_a) - 1
df2 = len(both_b) - 1
f_critical = st.f.ppf(1-0.05,df1,df2)
p_value = 1 - st.f.cdf(f_statistic,df1,df2)
if f_statistic < f_critical:
    print("The variances of the two groups are NOT significantly different.")
else:
    print("The variances of the two groups are significantly different.")
print(f"F-statistic: {f_statistic:.5f}")
print(f"Critical F-value: {f_critical:.5f}")
print(f"p-value: {p_value}")
```

# 8

Finally, test whether the mean of chlorophyll a is greater than mean of chlorophyll b using appropriate statistical test in all three combinations: Deciduous forests separately, Evergreen forests separately and both the forests together. Clearly state your null hypothesis, chosen significance criteria and the result of hypothesis testing.

For all three cases,

$H_0$: $\mu_a \leq \mu_b$

$H_1$: $\mu_a > \mu_b$

$\alpha = 0.05$

Since the chlorophyll a and chlorophyll b measurements in a given sample have been made on the same tree, they are paired. Hence, we must use the <b>paired $t$-test</b>.

```python
alpha = 0.05
```

<b>Deciduous forests</b>

```python
t_statistic, p_value = st.ttest_rel(a=deciduous_a,b=deciduous_b,alternative="greater")
if p_value > alpha:
    print("The mean of chlorophyll a is NOT significantly greater than that of chlorophyll b.")
else:
    print("The mean of chlorophyll a is significantly greater than that of chlorophyll b.")
df1 = len(deciduous_a) - 1
df2 = len(deciduous_b) - 1
t_critical = st.t.ppf(1-0.05,df1+df2)
print(f"t-statistic: {t_statistic:.5f}")
print(f"Critical t-value: {t_critical:.5f}")
print(f"p-value: {p_value}")
```

<b>Evergreen forests</b>

```python
t_statistic, p_value = st.ttest_rel(a=evergreen_a,b=evergreen_b,alternative="greater")
if p_value > alpha:
    print("The mean of chlorophyll a is NOT significantly greater than that of chlorophyll b.")
else:
    print("The mean of chlorophyll a is significantly greater than that of chlorophyll b.")
df1 = len(evergreen_a) - 1
df2 = len(evergreen_b) - 1
t_critical = st.t.ppf(1-0.05,df1+df2)
print(f"t-statistic: {t_statistic:.5f}")
print(f"Critical t-value: {t_critical:.5f}")
print(f"p-value: {p_value}")
```

<b>Both forests</b>

```python
t_statistic, p_value = st.ttest_rel(a=both_a,b=both_b,alternative="greater")
if p_value > alpha:
    print("The mean of chlorophyll a is NOT significantly greater than that of chlorophyll b.")
else:
    print("The mean of chlorophyll a is significantly greater than that of chlorophyll b.")
df1 = len(both_a) - 1
df2 = len(both_b) - 1
t_critical = st.t.ppf(1-0.05,df1+df2)
print(f"t-statistic: {t_statistic:.5f}")
print(f"Critical t-value: {t_critical:.5f}")
print(f"p-value: {p_value}")
```
