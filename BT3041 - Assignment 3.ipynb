```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset.csv").transpose()
```

```python
df.columns = df.iloc[0]
```

```python
df = df.drop("Gene",axis=0)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_df = scaler.fit_transform(df)
```

```python
scaled_df.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=20)
```

```python
pca.fit(scaled_df)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(dpi=200)
plt.bar(range(1,21),pca.explained_variance_ratio_)
plt.xticks(ticks=range(1,21),labels=range(1,21))
plt.title("Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Percentage of Explained Variance");
```

```python
transformed_df = pca.transform(scaled_df)
```

```python
transformed_df.shape
```

```python
transformed_df
```

```python
transformed_df = pd.DataFrame(data = transformed_df, columns=[f"PC {i}" for i in range(1,21)])
```

```python
transformed_df.index = df.index
```

```python
transformed_df
```

```python
metadata = pd.read_csv("Metadata_bt.csv")
```

```python
metadata["Sample"] = metadata["Sample"].apply(lambda x: x+".bam")
```

```python
metadata = metadata.set_index(metadata["Sample"]).drop("Sample",axis=1)
```

```python
metadata
```

```python
plt.figure(dpi=200)
sns.scatterplot(x="PC 1",y="PC 2",data=transformed_df,hue=metadata["Cell line/Tissue"])
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("PCA Plot");
```

```python
components = pd.DataFrame(abs(pca.components_)).transpose()
components.columns = [f"PC {i}" for i in range(1,21)]
components.index = df.columns
components
```
